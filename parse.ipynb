```python
import plotly.express as px

def get_map():
    map = px.scatter_map(
        stations,
        lat="lat",
        lon="lng",
        hover_name="name",
        hover_data=["name", "order"],
        # color_discrete_sequence=["gray"],
        color="hex_color",
        zoom=9,
        height=600,
        
    )

    # map.update_layout(locals="ru")
    # map.update_layout(map_style="open-street-map")
    
    map.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})

    return map
```
